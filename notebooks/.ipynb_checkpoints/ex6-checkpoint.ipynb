# Filtrando con un rechazabanda ideal

```python
def distance(x, y, cx, cy):
    return (x - cx)**2 + (y - cy)**2
```

```python
N, M = img.shape
img_ideal = np.zeros((N, M))
for i in range(N):
    for j in range(M):
        
        if (distance(i, j, 105, 172) <  15):
            img_spectrum[j, i] = 0
        
        if (distance(i, j, 190, 121) <  15):
            img_spectrum[j, i] = 0
        
        if (distance(i, j, 168, 158) <  15):
            img_spectrum[j, i] = 0
        
        if (distance(i, j, 125, 135) <  15):
            img_spectrum[j, i] = 0
```

```python
def notch_filter(x : np.array, cx, cy, gamma_h=0.4, gamma_l=1.6, c=1, D0=0.1, plot_filter=False):
    X = np.fft.fftshift(np.fft.fft2(x))
    
    d = np.array([np.sqrt((np.array(range(x.shape[0])) - cx)**2 + (i - cy)**2) for i in range(x.shape[1])]).T
    H = (gamma_h - gamma_l) * (1 - np.exp(-c * d**2 / (D0 * x.shape[0])**2)) + gamma_l
    H = np.interp(H, (H.min(), H.max()), (H.min(), 1))
    if (plot_filter):
        plt.imshow(H)
        plt.colorbar()

    Y = np.multiply(X, H.max() - H)
    y = np.real(np.fft.ifft2(np.fft.fftshift(Y)))
    return y, Y
```
